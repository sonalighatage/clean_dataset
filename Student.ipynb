```python
df = pd.read_csv(r"student-por.csv")


df.columns = df.columns.str.strip().str.lower()

print("Initial sex unique values:", df['sex'].unique())
print("Initial education unique values:", df['schoolsup'].unique())
```

```python
# Correctly map 'self_employed' after stripping whitespace from values
df['sex'] = df['sex'].str.strip().map({'F': 1, 'M': 0})

# Verify the unique values after correction
print(df['sex'].unique())
```

```python
x=df.drop(columns=['g3'],axis=1)
y=df['g3']
```

```python
x=df.drop(columns=['dalc'])
y=df['dalc']
```

```python
# train test split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
import pandas as pd

x_train_encoded = pd.get_dummies(x_train, drop_first=True)
x_test_encoded = pd.get_dummies(x_test, drop_first=True)

lr=LinearRegression()
lr.fit(x_train_encoded,y_train)
```

```python
y_train_pred = lr.predict(x_train_encoded)
y_test_pred = lr.predict(x_test_encoded)

#prediction on both train and test data
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score


mae=mean_absolute_error(y_train,y_train_pred)
mse=mean_squared_error(y_train,y_train_pred)
rmse=np.sqrt(mse)
r2=r2_score(y_train,y_train_pred)

print('mae',mae)
print('mse',mse)
print('rmse',rmse)
print('r2',r2)
```

```python
#rmse-it brings the error back to the original price unit ,
train_r2=r2_score(y_train,y_train_pred)
test_r2=r2_score(y_test,y_test_pred)
print('train_r2',train_r2)
print('test_r2',test_r2)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
# the reason for using mae,mse,rmse,r2 is that we neve judge a model by accuracy in LinaerRegression
# here we use error based matrics

mae = mean_absolute_error(y_test,y_test_pred)
mse = mean_squared_error(y_test,y_test_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test,y_test_pred)

print('MAE',mae)
print('MSE',mse)
print('RMSE',rmse)
print('R2',r2)
```
